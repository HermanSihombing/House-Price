```python
df_rumah = df_rumah.reset_index()
```

```python
feature_name = ['luas_tanah', 'luas_bangunan', 'jlh_kamar', 'toilet', 'lantai', 'carport']
X_data = df_rumah[feature_name]
y_data = df_rumah['harga']
```

```python
X_data.shape, y_data.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, random_state=0)

linreg = LinearRegression().fit(X_train, y_train)

print('linear model coeff (w): {}'.format(linreg.coef_))

print('linear model intercept (b): {:.3f}'.format(linreg.intercept_))

print('R-squared score (training): {:.3f}'.format(linreg.score(X_train, y_train)))

print('R-squared score (test): {:.3f}'.format(linreg.score(X_test, y_test)))
```

```python
predict_output = linreg.predict([[278, 350, 4, 4, 2, 2]])

print('Predict Harga Rumah 419 4 3 4 ',predict_output)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

bestfeatures = SelectKBest(score_func=chi2, k=3)
fit = bestfeatures.fit(X_data, y_data)
```

```python

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X_data.columns)
```

```python
featureScores = pd.concat([dfcolumns,dfscores], axis=1)
featureScores.columns = ['Specs', 'Scores']
```
